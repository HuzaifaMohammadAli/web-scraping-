```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import pandas as pd
from selenium.common.exceptions import NoSuchElementException
from time import sleep
import re


driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
```

```python
driver.get('https://www.daraz.pk')

input_search = driver.find_element(By.XPATH,"//input[@id='q']")
serach_button = driver.find_element(By.XPATH,"//button[@class='search-box__button--1oH7']")

# type any item to get the details
input_search.send_keys('oppo mobile')
sleep(2)
serach_button.click()
```

```python
data = []
for element in driver.find_elements(By.XPATH,"//div[@class='gridItem--Yd0sa']"):

# scrap the title
    title_element = element.find_element(By.XPATH,".//div[@class='title--wFj93']").text

# scrap price   
    price_element = element.find_element(By.XPATH,".//span[@class='currency--GVKjl']").text

# scrap numbers of reviews
    review_element = 0
    try:
        review_element = element.find_element(By.XPATH,".//span[@class='rating__review--ygkUy']").text
        review_element = int(''.join(filter(str.isdigit, review_element)))
    except NoSuchElementException:
        review_element = 0

# scrap rating 
    rating = 0
    try:
        element_rating = element.find_element(By.XPATH,".//div[@class='rating--ZI3Ol rate--DCc4j']").get_attribute('innerHTML')
        for s in  re.findall(r' star-(.*?)--', element_rating):
            rating = rating + int(s)
        rating = rating / 10
    except NoSuchElementException:
        rating = 0
  
    data.append([title_element , price_element, review_element, rating])
   
# Close the WebDriver when done
driver.quit()
```

```python
columns = ['Title','price','no_of_reviews','Rating']
df = pd.DataFrame(data=data , columns=columns)
df.head(10)
```
